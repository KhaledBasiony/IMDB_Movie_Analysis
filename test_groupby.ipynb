```python
import pandas as pd

df = pd.read_csv('IMDd-Movie-Database.csv', parse_dates = ['Release Date'])[['Release Date', 'Genre']]
df['Count'] = 1
df.head()
```

```python
df.set_index('Release Date').groupby(pd.Grouper(freq='M')).agg({'Genre': 'count'})
```

```python
table = df.pivot_table(index='Release Date', columns='Genre', values='Count', aggfunc='sum').resample('M').sum()
table
```
